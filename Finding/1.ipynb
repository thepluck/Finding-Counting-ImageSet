```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Read 1st image
img1 = cv2.imread("1.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
obj1 = img1[900:1150, 1500:1650]
plt.imshow(cv2.cvtColor(obj1, cv2.COLOR_BGR2RGB))
```

```python
img2 = img1.copy()
res = cv2.matchTemplate(img2, obj1, cv2.TM_CCOEFF_NORMED)
threshold = 0.6
loc = np.where(res >= threshold)
for pt in zip(*loc[::-1]):
    cv2.rectangle(img2, pt, (pt[0] + obj1.shape[1], pt[1] + obj1.shape[0]), (0, 0, 1), 2)
plt.imshow(img2, cmap='gray')
```
